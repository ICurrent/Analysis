```python
df = pd.read_csv('data')
df.head()
```

```python
df['Outstanding'] = df['Outstanding'].astype('str')
df['Above average'] = df['Above average'].astype('str')
df['Average'] = df['Average'].astype('str')
df['Below average'] = df['Below average'].astype('str')
df['Poor'] = df['Poor'].astype('str')
df['No opinion'] = df['No opinion'].astype('str')
df['Weighted average[54]'] = df['Weighted average[54]'].astype('str')
```

```python
df['Outstanding'] = df['Outstanding'].str.replace('%', '')
df['Above average'] = df['Above average'].str.replace('%', '')
df['Average'] = df['Average'].str.replace('%', '')
df['Below average'] = df['Below average'].str.replace('%', '')
df['Poor'] = df['Poor'].str.replace('%', '')
df['No opinion'] = df['No opinion'].str.replace('%', '')
df['Weighted average[54]'] = df['Weighted average[54]'].str.replace('%', '')
```

```python
df.head()
```

```python
df['Outstanding'] = df['Outstanding'].astype('float')
df['Above average'] = df['Above average'].astype('float')
df['Average'] = df['Average'].astype('float')
df['Below average'] = df['Below average'].astype('float')
df['Poor'] = df['Poor'].astype('float')
df['No opinion'] = df['No opinion'].astype('float')
df['Weighted average[54]'] = df['Weighted average[54]'].astype('float')
```

```python
df.head()
```

```python
# create plot

fig, ax = plt.subplots(figsize=(8,8), dpi = 250)
plot = sns.heatmap(df.iloc[:, 1:], annot= True, cmap='RdBu_r', ax=ax)

plot.set_yticklabels(df['President'], rotation = 360, fontsize = 8)
plot.set_xticklabels(plot.get_xticklabels(), rotation = 45, fontsize = 8, ha = 'right')
plt.title('Presidential Ranking');
```
