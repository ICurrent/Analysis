```python
import mysql.connector
import pandas as pd
from timeit import default_timer as timer
```

```python
database = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = '',
    database = 'W3schools'
)
```

```python
cursor = database.cursor()
cursor.execute('SHOW TABLES')

table = []

for i in cursor:
    table.append(i[0])
print(table)
    #break
```

```python
for i in table:
    data = pd.read_sql('''
        SELECT * 
        FROM {}
    '''.format(i), database)
    data.to_csv(i, index=False)
```

```python
pd.read_sql('''
SELECT c.CustomerName, c.City, ord.OrderID
FROM customers c, orders ord
''', database)
```

```python
pd.read_sql('''
SELECT ord.OrderID, c.CustomerName, 
	GROUP_CONCAT(p.ProductName,' - ', ord.Quantity SEPARATOR '\n') ProductPerQty,
    SUM(ord.Quantity) TotalQty
    FROM order_details ord
    JOIN products p ON ord.ProductID = p.ProductID
    JOIN orders od ON ord.OrderID = od.OrderID
    JOIN customers c ON od.CustomerID = c.CustomerID
    GROUP BY 1
    ORDER BY 2
    LIMIT 10;
''', database)
```
